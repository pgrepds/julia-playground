## $f(x)=e^x$

```julia
f(x) = exp(x)
```

The above function $f(x)=e^x$ is the famous exponential function, whereby $e$ refers to Euler's number.

```julia
for x in -5:5
    println(f(x))
end
```

```julia
using Plots
plot(f, -5, 5)
plot!(title = "The exponential function", leg=false)
```
