```julia
using Pkg
Pkg.add("Plots")
```

```julia
using Plots
```

```julia
plot(
    sin,
    -2π, 2π,
)
```

```julia
plot(
    x -> -log(x),
    0, 2000,
    label="-log(x)",
    xlabel="x",
    ylabel="y=-log(x)",
    title="Title for plot",
)
```

```julia
# create scatter plots
x = [1:20]
y1, y2 = rand(20), rand(20)
plot(x, [y1, y2], seriestype=:scatter, thickness_scaling=0.6, opacity=0.8)
```
